```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('car_crashes')
```

```python
data.head(10)
```

## Load the data you receive into a Pandas

```python
X = data["total"].values
y = data["speeding"].values
z = data["alcohol"].values
w = data["not_distracted"].values
f = data["abbrev"].values

d = {'total': X, 'speeding': y, 'alcohol': z, 'not_distracted': w, 'abbrev': f}

df = pd.DataFrame(data=d)
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(20, 13))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="abbrev", y="total",
                hue="speeding", size="alcohol", sizes=(50, 500), 
                data=df, ax=ax)
```

```python
f, ax = plt.subplots(figsize=(20, 13))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="abbrev", y="alcohol",
                hue="speeding", size="not_distracted", sizes=(50, 500),
                data=df, ax=ax)
```
