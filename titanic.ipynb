# Data Visualization - titanic dataset

```python
import seaborn as sns
import pandas as pd
```

```python
sns.set()
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('titanic')
```

```python
data.head(10)
```

## Load the data you receive into a Pandas

```python
X = data["age"].values
y = data["fare"].values
z = data["survived"].values
w = data["who"].values
f = data["embark_town"].values
d = data["alive"].values
e = data["class"].values
g = data["sex"].values

d = {'age': X, 'fare': y, 'survived': z, 'who': w, 'embark_town': f, 'alive': d, 'class': e, 'sex': g}

df = pd.DataFrame(data=d)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="age", y="fare",
                hue="survived", size="who",
                data=df, ax=ax)
```

```python
sns.set_theme(style="whitegrid")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=df, x="age", y="embark_town", hue="alive",
)
sns.despine(left=True)
```

```python
sns.set_theme(style="white")
grouped_by_sex=df.groupby('sex')
df
sns.relplot(x="age", y='fare', hue="alive", size="class", data=df)
```
